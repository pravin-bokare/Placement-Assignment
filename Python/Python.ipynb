# Q2. Consider a string to be valid if all characters of the string appear the same number of times. It is also valid if he can remove just one character at the index in the string, and the remaining characters will occur the same number of times. Given a string, determine if it is valid. If so, return YES , otherwise return NO .

```python
def valid(string):
    words = list(string)
    a = words.count(words[0])
    for i in words:
        if words.count(i)!=a:
            return 'NO'
    return 'YES'

valid('abc')
```

# Q4. Write a program to download the data from the link given below and then read the data and convert the into the proper structure and return it as a CSV file.

```python
import pandas as pd
import requests

def download_and_convert_to_csv(link):
    # Download the data from the provided link
    response = requests.get(link)
    data = response.json()

    # Extract the required attributes from the data
    structured_data = []
    for meteorite in data:
        try:
            meteorite_data = {
                "name": meteorite.get("name"),
                "id": meteorite.get("id"),
                "nametype": meteorite.get("nametype"),
                "recclass": meteorite.get("recclass"),
                "mass": meteorite.get("mass"),
                "year": meteorite.get("year"),
                "reclat": meteorite.get("reclat"),
                "reclong": meteorite.get("reclong")
            }
            structured_data.append(meteorite_data)
        except KeyError as e:
            print(f"KeyError: {e} - Skipping meteorite")

    # Convert the structured data to a DataFrame
    df = pd.DataFrame(structured_data)

    # Convert the DataFrame to a CSV file
    csv_file = "earth_meteorites.csv"
    df.to_csv(csv_file, index=False)

    return csv_file

# Example usage
input_link = "https://data.nasa.gov/resource/y77d-th95.json"
output_csv = download_and_convert_to_csv(input_link)
print("Data downloaded and converted to CSV:", output_csv)

# Load the data from the CSV file
data = pd.read_csv("earth_meteorites.csv")
data.head()
```

# Q5. Write a program to download the data from the given API link and then extract the following data with proper formatting

```python
import requests
import pandas as pd
import re

def download_and_extract_data(link):
    # Download the data from the provided API link
    response = requests.get(link)
    data = response.json()

    # Extract the required attributes from the data
    show_id = data["id"]
    show_url = data["url"]
    show_name = data["name"]

    episodes = data["_embedded"]["episodes"]
    extracted_data = []
    for episode in episodes:
        try:
            episode_data = {
                "id": episode["id"],
                "url": episode["url"],
                "name": episode["name"],
                "season": episode["season"],
                "number": episode["number"],
                "type": episode["type"],
                "airdate": episode["airdate"],
                "airtime": episode["airtime"],
                "runtime": episode["runtime"],
                "average_rating": episode["rating"]["average"],
                "summary": re.sub('<[^<]+?>', '', episode["summary"]),
                "medium_image_link": episode["image"]["medium"],
                "original_image_link": episode["image"]["original"]
            }
            extracted_data.append(episode_data)
        except:
            continue

    # Convert the structured data to a DataFrame
    df = pd.DataFrame(extracted_data)

    # Convert the DataFrame to a CSV file
    csv_file = "episode_output.csv"
    df.to_csv(csv_file, index=False)

    return csv_file

input_link = "http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes"
output_csv = download_and_extract_data(input_link)
print("Data downloaded and converted to CSV")

df = pd.read_csv("episode_output.csv")
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv("episode_output.csv")

# Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.
season_ratings = df.groupby("season")["average_rating"].mean().reset_index()
plt.bar(season_ratings["season"], season_ratings["average_rating"])
plt.xlabel("Season")
plt.ylabel("Average Rating")
plt.title("Ratings for Each Season")
plt.show()
```

```python
# Get all the episode names, whose average rating is more than 8 for every season
high_rating_episodes = df[df["average_rating"] > 8]
high_rating_episodes = high_rating_episodes.groupby("season")["name"].apply(list).reset_index()
for index, row in high_rating_episodes.iterrows():
    season = row["season"]
    episodes = row["name"]
    print(f"Season {season} - High Rating Episodes: {episodes}")
```

```python
# Get all the episode names that aired before May 2019
df["airdate"] = pd.to_datetime(df["airdate"])
aired_before_2019 = df[df["airdate"] < pd.Timestamp(2019, 5, 1)]
aired_before_2019_episode_names = aired_before_2019["name"].tolist()
print("Episodes Aired Before May 2019:")
aired_before_2019_episode_names
```

```python
# Get the episode name from each season with the highest and lowest rating
max_rating_episodes = df.groupby("season").apply(lambda x: x[x["average_rating"] == x["average_rating"].max()])["name"]
min_rating_episodes = df.groupby("season").apply(lambda x: x[x["average_rating"] == x["average_rating"].min()])["name"]
rating_comparison = pd.concat([max_rating_episodes, min_rating_episodes], axis=1)
rating_comparison.columns = ["Highest Rating Episode", "Lowest Rating Episode"]
print("Episode with Highest and Lowest Rating for Each Season:")
rating_comparison.head()
```

```python
# Get the summary for the most popular ( ratings ) episode in every season
most_popular_episodes = df.groupby("season").apply(lambda x: x[x["average_rating"] == x["average_rating"].max()])["summary"]
print("Summary for the Most Popular Episode in Every Season:")
most_popular_episodes
```

# Q10. Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary.
Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide
an explanation for the same.

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def count_pos(text):
    # Tokenize the text into words
    words = word_tokenize(text)
    
    # Perform part-of-speech tagging
    tagged_words = pos_tag(words)
    
    # Initialize counts
    counts = {'verbs': 0, 'nouns': 0, 'pronouns': 0, 'adjectives': 0}
    
    # Count the occurrences of each POS category
    for word, tag in tagged_words:
        if tag.startswith('VB'):
            counts['verbs'] += 1
        elif tag.startswith('NN'):
            counts['nouns'] += 1
        elif tag.startswith('PRP'):
            counts['pronouns'] += 1
        elif tag.startswith('JJ'):
            counts['adjectives'] += 1
    
    return counts

# Test case 1
text1 = "The quick brown fox jumps over the lazy dog."
result1 = count_pos(text1)
print(result1)  # Output: {'verbs': 1, 'nouns': 5, 'pronouns': 0, 'adjectives': 2}

# Test case 2
text2 = "She sells seashells by the seashore."
result2 = count_pos(text2)
print(result2)  # Output: {'verbs': 1, 'nouns': 3, 'pronouns': 1, 'adjectives': 0}

# Additional test case 1
text3 = "I love eating delicious pizza."
result3 = count_pos(text3)
print(result3)  # Output: {'verbs': 1, 'nouns': 2, 'pronouns': 0, 'adjectives': 1}

# Additional test case 2
text4 = "They are playing a friendly football match."
result4 = count_pos(text4)
print(result4)  # Output: {'verbs': 2, 'nouns': 3, 'pronouns': 1, 'adjectives': 1}
```
